```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(357823)
```

```python
dataframe = pd.read_csv('movies.csv', delimiter = ';')
#Esto quita las filas con valores null
dataframe.dropna(inplace=True)
dataframe.head()
```

```python
#Obtener todos los géneros de las películas 
dataframe["genres"].str.split("-").explode().unique()
```

```python
len(dataframe)
```

```python
#Modificar el dataframe de forma que tenga columnas binarias por cada género 
# y eliminando la columna de géneros (ya no es relevante)

dataframe = dataframe.drop(columns="genres").join(dataframe["genres"].str.get_dummies(sep="-"))
dataframe.head()
```

```python
#Pasar a csv el dataframe nuevo
dataframe.to_csv('movies_preprocesado.csv', index=False)
```
